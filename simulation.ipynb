# Simulation

```python
import numpy as np
import physics

```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 500  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = 0.01  # s
time_max = 10  # s
```

```python
time = 0  # s
depth = 0  # m
velocity = 0  # m/s
acceleration = 0  # m/s^2
buoyancy =  density_water * object_volume * g # N
weight =  object_volume * density_object * g # N
```

```python
# lists for plotting later
time = np.arange(0, time_max, time_step)
position = np.zeros_like(time)
velocity = np.zeros_like(time)
acceleration = np.zeros_like(time)
```

```python
# simulation loop
velocity_old = 0
position_old = 0
for i in range(1, len(time)):
    # calculate the force on the object
    force = buoyancy - weight
    # calculate acceleration
    acceleration[i] = force / (object_volume * density_object)
    # calculate velocity
    velocity[i] = acceleration[i]*time[i] + velocity[0]
    # calculate position
    position[i] = (velocity[i]**2-velocity[0]**2)/(2*acceleration[i]) 
    pass
```

# Plot the results

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(time, position, label="Position")
plt.plot(time, velocity, label="Velocity")
plt.plot(time, acceleration, label="Acceleration")
plt.xlabel("Time (s)")
plt.ylabel("Position (m), Velocity (m/s), Acceleration (m/s^2)")
plt.legend()
plt.show()
```
